```python
DNA  = 'ATGTCCTTTTTATCG'
complement = {
    'A':'T',
    'T':'A',
    'C':'G',
    'G':'C'
}

complement = ''.join(complement[i] for i in DNA)
print("DNA Sequence: ",DNA)
print("Complement of DNA sequence: ",complement)
```

```python
mRNA = complement.replace('T','U')
print("mRNA sequence: ",mRNA)
```

```python
codons = [mRNA[i:i+3] for i in range (0,len(mRNA),3)]
print("Codons: ",codons)
```

```python
def translate_to_amino_acids(codons):
    amino_acids = "".join(codon_table.get(codon,"?") for codon in codons)
    return amino_acids

codon_table = {
    "UUU": "F", 'UUC': 'F', 'UUA': 'L', 'UUG':'L',
    'CUU': 'L', 'CUC': 'L', 'CUA': 'L', 'CUG': 'L',
    'AUU': 'I', 'AUC': 'I', 'AUA': 'I','AUG': 'M',
    'GUU': 'V', 'GUC': 'V', 'GUA': 'V', 'GUG': 'V',
    'UCU': 'S', 'UCC': 'S', 'UCA': 'S', 'UCG': 'S', 
    'AGU': 'S', 'AGC': 'S',
    'CCU': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P',
    'ACU': 'T', 'ACC': 'T', 'ACA': 'T', 'ACG': 'T',
    'GCU': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A',
    'UAU': 'Y', 'UAC': 'Y',
    'CAU': 'H', 'CAC': 'H',
    'CAA': 'Q', 'CAG': 'Q',
    'AAU': 'N', 'AAC': 'N',
    'AAA': 'K', 'AAG': 'K',
    'GAU': 'D', 'GAC': 'D',
    'GAA': 'E', 'GAG': 'E',
    'UGU': 'C', 'UGC': 'C',
    'UGG': 'W',
    'CGU': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R', 'AGA': 'R', 'AGG': 'R',
    'GGU': 'G', 'GGC': 'G', 'GGA': 'G', 'GGG': 'G',
    'UAA': 'Stop', 'UAG': 'Stop', 'UGA': 'Stop'
}

#Codons =  ['UAC', 'AGG', 'AAA', 'AAU', 'AGC']
amino_acid_seq = translate_to_amino_acids(codons)
print("Amino acid sequence: ",amino_acid_seq)
```
