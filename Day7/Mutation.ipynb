```python
#2nd method for find complement

def complementary_strand(dna_seq):
    complement_map = str.maketrans("ATGC","TACG")
    complement = dna_seq.translate(complement_map)
    return complement

dna = "ATGTCCTTTTTATCG"
complement = complementary_strand(dna)

print("Original DNA sequence:       ", dna)
print("Complement sequence:         ", complement)

def dna_to_mrna(dna_seq):
    mrna = complement.replace("T","U")
    return mrna

mrna = dna_to_mrna(complement)
print("mRNA sequence:       ", mrna)
```

```python
def split_into_codons(mrna_seq):
    codons = [mrna_seq[i:i+3] for i in
             range(0,len(mrna_seq),3)
             if len(mrna_seq[i:i+3]) == 3]
    return codons

codons = split_into_codons(mrna)
print("Codons: ", codons)
```

```python
codon_table = {
    "UUU": "F", 'UUC': 'F', 'UUA': 'L', 'UUG':'L',
    'CUU': 'L', 'CUC': 'L', 'CUA': 'L', 'CUG': 'L',
    'AUU': 'I', 'AUC': 'I', 'AUA': 'I','AUG': 'M',
    'GUU': 'V', 'GUC': 'V', 'GUA': 'V', 'GUG': 'V',
    'UCU': 'S', 'UCC': 'S', 'UCA': 'S', 'UCG': 'S', 
    'AGU': 'S', 'AGC': 'S',
    'CCU': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P',
    'ACU': 'T', 'ACC': 'T', 'ACA': 'T', 'ACG': 'T',
    'GCU': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A',
    'UAU': 'Y', 'UAC': 'Y',
    'CAU': 'H', 'CAC': 'H',
    'CAA': 'Q', 'CAG': 'Q',
    'AAU': 'N', 'AAC': 'N',
    'AAA': 'K', 'AAG': 'K',
    'GAU': 'D', 'GAC': 'D',
    'GAA': 'E', 'GAG': 'E',
    'UGU': 'C', 'UGC': 'C',
    'UGG': 'W',
    'CGU': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R', 'AGA': 'R', 'AGG': 'R',
    'GGU': 'G', 'GGC': 'G', 'GGA': 'G', 'GGG': 'G',
    'UAA': 'Stop', 'UAG': 'Stop', 'UGA': 'Stop'
}

def translate_to_amino_acids(codons):
    amino_acids = "".join(codon_table.get(codons,"?")
                         for codon in codons)
    return amino_acids

amino_acid_seq = translate_to_amino_acids(codons)
print("Amino acid sequence: ",amino_acid_seq)
```
