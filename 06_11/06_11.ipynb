```python
#Q1

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#load dataset
data = pd.read_csv("dea_data.csv")
```

```python
#Q2

data.head(5)
```

```python
#Q3

control_cols = [i for i in data.columns if "Control" in i]
print("\n")
print("Colums of Controls: ")
print(control_cols)
```

```python
#Q3

treatment_cols = [i for i in data.columns if "Treatment" in i]
print("\n")
print("Colums of Treatment: ")
print(treatment_cols)
```

```python
gene = "Gene12"
gene_data = data[data["Gene"] == gene]
control_mean = gene_data[control_cols].mean(axis = 1).values[0]
print("Average Expression Level: ",control_mean)
```

```python
treatment_mean = gene_data[treatment_cols].mean(axis = 1).values[0]
print("Average Expression Level: ",treatment_mean)
```

```python
#Q4

for gene in data["Gene"].head(100):
    gene_data = data[data["Gene"]==gene]
    
    control_mean=gene_data[control_cols].mean(axis = 1).values[0]
    print("Gene: ",gene)
    print("Average Expression Level(under control): ",control_mean)
```

```python
#Q5

for gene in data["Gene"].head(100):
    gene_data = data[data["Gene"]==gene]
    
    treatment_mean = gene_data[treatment_cols].mean(axis = 1).values[0]
    print("Gene: ",gene)
    print("Average Expression Level(under treatment): ",treatment_mean)
```

```python
#Q6

data["Average_treatment"]=data[treatment_cols].mean(axis=1)

highest_gene = data.loc[data['Average_treatment'].idxmax()]

print("Gene:" ,highest_gene['Gene'])
print("Highest Average Expression Level: ",highest_gene['Average_treatment'])
```

```python
#Q7

data["Average_treatment"]=data[treatment_cols].mean(axis=1)
data["Average_control"]=data[control_cols].mean(axis=1)

data['Difference'] = data['Average_treatment'] - data['Average_control']
print(data[['Gene', 'Average_treatment', 'Average_control','Difference']])
```

```python
#Q8

gene = "Gene10"
gene_data = data[data["Gene"] == gene]

control_mean = gene_data[control_cols].mean(axis = 1).values[0]
treatment_mean = gene_data[treatment_cols].mean(axis = 1).values[0]

plt.bar("Control", control_mean,color='blue', label = "Control")
plt.bar("Treatment", treatment_mean,color='green', label = "Treatment")
plt.title(f"Bar chart: {gene} Average Expression")
plt.ylabel("Expression level")
plt.legend()
plt.show()
```

```python
#Q9

heatmap_data = data.set_index('Gene')
heatmap =  heatmap_data.head(100)

plt.figure(figsize = (12,8))
sns.heatmap(heatmap, cmap= "coolwarm", annot= True)
plt.title("Headmap of Gene expression Levels")
plt.xlabel("Samples")
plt.ylabel("Genes")
plt.show()
```

```python
#Q10 Identify gene that are upregulated (treatment_avg > control_avg)

data["Average_treatment"]=data[treatment_cols].mean(axis=1)
data["Average_control"]=data[control_cols].mean(axis=1)

data['Difference'] = data['Average_treatment'] - data['Average_control']

upregulated = data[data['Difference'] > 0]

print("Upregulated Genes: ")
print(upregulated[['Gene', 'Average_treatment', 'Average_control','Difference' ]])
```

```python
#Q10 Identify gene that are downregulated (treatment_avg < control_avg)

data["Average_treatment"]=data[treatment_cols].mean(axis=1)
data["Average_control"]=data[control_cols].mean(axis=1)

data['Difference'] = data['Average_treatment'] - data['Average_control']

upregulated = data[data['Difference'] < 0]

print("Downregulated Genes: ")
print(upregulated[['Gene', 'Average_treatment', 'Average_control','Difference' ]])
```

```python
#Q12 Box plot of the treatement expression levels for the first 5 genes

treatment_data = data[['Treatment1', 'Treatment2', 'Treatment3', 'Treatment4', 'Treatment5']].values.flatten()

plt.figure(figsize=(8,6))
sns.boxplot(data = [treatment_data], palette = "Set2")

plt.xticks([1],['Treatment'])
plt.title(f"Box plot for first 5 gene")
plt.ylabel("Treatment Expression level")

plt.show()
```

```python
#Q13 calculate overall mean expression 

overall_treatment_mean = data[treatment_cols].mean().mean()
overall_control_mean = data[control_cols].mean().mean()

print("Overall mean expression levels:")
print("Treatment samples: ",overall_treatment_mean)
print("Control samples: ",overall_control_mean)
```

```python
#Q14 Greatest overall variation
all_samples = control_cols + treatment_cols
data['Variation'] = data[all_samples].std(axis = 1)

max_var_gene=data.loc[data['Variation'].idxmax()]
print("Gene with Greatest overall variation across all samples:")
print("Gene: ",max_var_gene['Gene'])
print('Variation: ',max_var_gene['Variation'])
```
