```python
#DNA sequence = ATGCCGTAGCTAAGGCTT 
#using this DNA sequence create,
#A subsitition mutation 
#An insertion mutation 
#A deletion mutation
```

```python
original_dna = "ATGCCGTAGCTAAGGCTT"

# 1. Substitution Mutation
# Replace the 5th base (index 4) with 'A'
substitution_dna = original_dna[:4] + 'A' + original_dna[5:]

# 2. Insertion Mutation
# Insert 'T' after the 6th base (index 6)
insertion_dna = original_dna[:6] + 'T' + original_dna[6:]

# 3. Deletion Mutation
# Remove the 8th base (index 7)
deletion_dna = original_dna[:7] + original_dna[8:]

# Print results
print("Original DNA:     ", original_dna)
print("Substitution:     ", substitution_dna)
print("Insertion:        ", insertion_dna)
print("Deletion:         ", deletion_dna)
```

```python
#transalate all 4 sequences into protein

#perform transcription and translation using the codon table
def complementary_strand(dna_seq):
    complement_map = str.maketrans("ATGC","TACG")
    complement = dna_seq.translate(complement_map)
    return complement

complement1 = complementary_strand(original_dna)
complement2 = complementary_strand(substitution_dna)
complement3 = complementary_strand(insertion_dna)
complement4 = complementary_strand(deletion_dna)

#create mRNA sequence
def dna_to_mrna(dna_seq):
    mrna = dna_seq.replace("T","U")
    return mrna

mrna1 = dna_to_mrna(complement1)
mrna2 = dna_to_mrna(complement2)
mrna4 = dna_to_mrna(complement3)
mrna3 = dna_to_mrna(complement4)

def split_into_codons(mrna_seq):
    codons = [mrna_seq[i:i+3] for i in
             range(0,len(mrna_seq),3)
             if len(mrna_seq[i:i+3]) == 3]
    return codons

codons1 = split_into_codons(mrna1)
codons2 = split_into_codons(mrna2)
codons3 = split_into_codons(mrna3)
codons4 = split_into_codons(mrna4)

print("Original DNA sequence:       ", original_dna)
print("Complement sequence:         ", complement1)
print("mRNA sequence:       ", mrna1)
print("Codons: ", codons1)

print("\n")

print("Substitution DNA sequence:         ", substitution_dna)
print("Complement sequence:         ", complement2)
print("mRNA sequence:       ", mrna2)
print("Codons: ", codons2)

print("\n")

print("Insertion DNA sequence:       ", insertion_dna)
print("Complement sequence:         ", complement3)
print("mRNA sequence:       ", mrna3)
print("Codons: ", codons3)

print("\n")

print("Deletion DNA sequence:       ", deletion_dna)
print("Complement sequence:         ", complement4)
print("mRNA sequence:       ", mrna4)
print("Codons: ", codons4)
```

```python
def translate_to_amino_acids(codons):
    amino_acids = "".join(codon_table.get(codon,"?") for codon in codons)
    return amino_acids

codon_table = {
    "UUU": "F", 'UUC': 'F', 'UUA': 'L', 'UUG':'L',
    'CUU': 'L', 'CUC': 'L', 'CUA': 'L', 'CUG': 'L',
    'AUU': 'I', 'AUC': 'I', 'AUA': 'I','AUG': 'M',
    'GUU': 'V', 'GUC': 'V', 'GUA': 'V', 'GUG': 'V',
    'UCU': 'S', 'UCC': 'S', 'UCA': 'S', 'UCG': 'S', 
    'AGU': 'S', 'AGC': 'S',
    'CCU': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P',
    'ACU': 'T', 'ACC': 'T', 'ACA': 'T', 'ACG': 'T',
    'GCU': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A',
    'UAU': 'Y', 'UAC': 'Y',
    'CAU': 'H', 'CAC': 'H',
    'CAA': 'Q', 'CAG': 'Q',
    'AAU': 'N', 'AAC': 'N',
    'AAA': 'K', 'AAG': 'K',
    'GAU': 'D', 'GAC': 'D',
    'GAA': 'E', 'GAG': 'E',
    'UGU': 'C', 'UGC': 'C',
    'UGG': 'W',
    'CGU': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R', 'AGA': 'R', 'AGG': 'R',
    'GGU': 'G', 'GGC': 'G', 'GGA': 'G', 'GGG': 'G',
    'UAA': 'Stop', 'UAG': 'Stop', 'UGA': 'Stop'
}

#Codons =  ['UAC', 'AGG', 'AAA', 'AAU', 'AGC']
amino_acid_seq1 = translate_to_amino_acids(codons1)
amino_acid_seq2= translate_to_amino_acids(codons2)
amino_acid_seq3 = translate_to_amino_acids(codons3)
amino_acid_seq4= translate_to_amino_acids(codons4)

print("Amino acid sequence of original DNA: ",amino_acid_seq1)
print("Amino acid sequence of subsitution Mutation: ",amino_acid_seq2)
print("Amino acid sequence of insertion mutation: ",amino_acid_seq3)
print("Amino acid sequence: of deletion mutation",amino_acid_seq4)
```

```python


# Function to calculate identity percentage
def calculate_identity(seq1, seq2):
    min_len = min(len(seq1), len(seq2))
    exact_matches = sum(1 for a, b in zip(seq1[:min_len], seq2[:min_len]) if a == b)
    identity_percentage = (exact_matches / min_len) * 100
    return exact_matches, min_len, identity_percentage

# Compare each mutation
mutations = {
    "Substitution": substitution_dna,
    "Insertion": insertion_dna,
    "Deletion": deletion_dna
}

print("DNA Identity Comparison:\n")
for label, mutated_seq in mutations.items():
    matches, positions, percent = calculate_identity(original_dna, mutated_seq)
    print(f"{label:<13}: {matches} exact matches out of {positions} positions → {percent:.2f}% identity")
```
